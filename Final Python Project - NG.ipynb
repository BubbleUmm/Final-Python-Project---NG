```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

### 1. Create a visual showing the relationship between horsepower and fuel economy."

```python
sns.scatterplot(data=mpg, x="horsepower", y="mpg", hue="origin", palette = "plasma")
plt.title("Horsepower vs Fuel Economy")
plt.xlabel("Horsepower")
plt.ylabel("MPG(Fuel Economy)")
```

#### Looking at this data, we can determine that as MPG(Fuel Economy) is increases, the Horsepower decreases. 
#### Meaning that there is an inverse relationship between horsepower and fuel economy.

### 2. Create a table summarizing fuel economy by country of origin

```python
sns.barplot(data=mpg, x="origin", y="mpg",  errorbar=None, hue="origin", palette = "plasma")
plt.title("Fuel Economy by Country of Origin")
plt.xlabel("Country of Origin")
plt.ylabel("MPG(Fuel Economy)")
```

#### Using this information provided, we can concluded among, that USA has the lowest MPG(Fuel Economy), 
#### followed by Europe, and then Japan. 
#### Therefore, the USA is the region with the least fuel efficiency vehicles.

### 3. Create a visual showing how fuel economy has changed over time

```python
sns.lineplot(data=mpg, x="model_year", y="mpg", hue="origin", palette = "plasma")
plt.title("Fuel Economy Change Over Time")
plt.xlabel("Model_Year")
plt.ylabel("MPG(Fuel Economy)")
```

#### As shown on this graph, the change Fuel Economy (measured in MPG) shows a increase over time, 
#### when measured over model_years as it progressed.
#### This means that vehichles became more fuel efficient as MPG values increased. 

#### The graphs presents a trend of improvement in the fuel economy accross the years shown, 
#### which is reflecting that vehichles have increased efficiency because of advancements over time.

### 4. Create two other visuals or tables related to fuel economy.

```python
sns.countplot(data=mpg, x="origin", hue="origin", palette = "plasma")
plt.title("Vehichle Count vs Country of Origin")
plt.xlabel("Country of Origin")
plt.ylabel("Count")
```

#### The bar chart shows the distribution of vehichles by "Country of Origin". 
#### The USA has the highest count, surpassing Japan and Europe,
#### which as shown have somewhat similar counts.
#### But overall, USA has the highest data of vehichles in the dataset.

```python
avg_mpg_by_origin = mpg.groupby("origin")["mpg"].mean().reset_index()
sns.barplot(data=avg_mpg_by_origin, x="origin", y="mpg", hue="origin", palette = "plasma")
plt.title("Avg. Fuel Economy by Country of Origin")
plt.xlabel("Country of Origin")
plt.ylabel("Average of MPG")
```

#### This bar chart compares avergae MPG by Country of Origin.
#### Japan has the lead in fuel efficiency, followed by Europe, and then the USA having the lowest MPG.
